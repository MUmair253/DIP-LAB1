```python
import matplotlib.pyplot as plt
import cv2

try:
    img = cv2.imread("C:\\Users\\abc\\Desktop\\Compressed\\Umair\\HV-ALPSRP247640500-H2.2_UA.tif") 
    (height, width) = img.shape[:2]    
    res = cv2.resize(img, (int(width / 4), int(height / 4)), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('result.tiff', res)
    plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))  # Convert color channels for Matplotlib
    plt.show()
 
except IOError:
    print('Error while reading files !!!')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

M = np.float32([[1, 0, 200], [0, 1, 200]])

try:
    img = cv2.imread("C:\\Users\\abc\\Desktop\\Compressed\\Umair\\HV-ALPSRP247640500-H2.2_UA.tif")
    rows, cols = img.shape[:2]
    
    res2 = cv2.warpAffine(img, M, (cols, rows))
    cv2.imwrite('result_shifted.tiff', res2)
    
    plt.imshow(cv2.cvtColor(res2, cv2.COLOR_BGR2RGB))  # Convert color channels for Matplotlib
    plt.show()

except IOError:
    print('Error while reading files !!!')
```

```python
import cv2
import matplotlib.pyplot as plt

try:
    img = cv2.imread("C:\\Users\\abc\\Desktop\\Compressed\\Umair\\HV-ALPSRP247640500-H2.2_UA.tif")
    edges = cv2.Canny(img, 50, 100)
    cv2.imwrite('result_edges.tiff', edges)
    
    plt.imshow(edges, cmap='gray')  # Display the edges in grayscale
    plt.show()

except IOError:
    print('Error while reading files !!!')
```
